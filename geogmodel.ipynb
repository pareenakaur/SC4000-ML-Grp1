### Aggregate rows 

```python
df = pd.read_csv('Geog_complete_train.csv')
df['same_city'] = (df['city_id'] == df['city_id_merchant']).astype(int)  # 1 if same, 0 otherwise
df['same_state'] = (df['state_id']==df['state_id_merchant']).astype(int)
```

Not all of these aggregated columns are used in the train set in the end. Mean and max for categorical variables are dropped. Refer to train_X.csv and train_y.csv for what columns are passed to the models 

```python

from scipy import stats  

PureGeog_grouped = df.groupby('card_id').agg({
    'city_id_merchant': ['mean', lambda x: stats.mode(x, keepdims=True)[0][0], 'nunique'],  # Mode, Mean, Unique Count
    'state_id_merchant': ['mean', lambda x: stats.mode(x, keepdims=True)[0][0], 'nunique'],  
    'city_id': ['mean', lambda x: stats.mode(x, keepdims=True)[0][0], 'nunique'],
    'state_id': ['mean', lambda x: stats.mode(x, keepdims=True)[0][0], 'nunique'], 
    'feature_1': ['mean', 'max', lambda x: stats.mode(x, keepdims=True)[0][0]],  
    'feature_2': ['mean', 'max', lambda x: stats.mode(x, keepdims=True)[0][0]],  
    'feature_3': ['mean', 'max', lambda x: stats.mode(x, keepdims=True)[0][0]],  
    'target': ['mean', 'max'],  
    'same_city': ['sum', 'mean'],  # Sum gives count of same city occurrences, mean gives % of same city transactions
    'same_state': ['sum', 'mean'], 
}).reset_index()
```

### Catboost

```python
import pandas as pd
import catboost as cb

# Categorical columns
categorical_cols = ['city_id_merchant_mode', 'state_id_merchant_mode', 'city_id_mode', 'state_id_mode', 'feature_1_mode', 'feature_2_mode', 'feature_3_mode']  

# Convert categorical columns to category dtype (recommended for efficiency)
X_train[categorical_cols] = X_train[categorical_cols].astype('category')
X_test[categorical_cols] = X_test[categorical_cols].astype('category')

# Define CatBoost model
catboost_model = cb.CatBoostRegressor(
    iterations=500, 
    learning_rate=0.1, 
    depth=6, 
    verbose=100
)

# Train the model, specifying categorical features
catboost_model.fit(X_train, y_train, cat_features=categorical_cols)

# Make predictions
y_pred = catboost_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
r2 = r2_score(y_test, y_pred)

# Save metrics to a file
'''with open("model_metrics.txt", "w") as file:
    file.write(f"Mean Absolute Error (MAE): {mae}\n")
    file.write(f"Mean Squared Error (MSE): {mse}\n")
    file.write(f"Root Mean Squared Error (RMSE): {rmse}\n")
    file.write(f"R² Score: {r2}\n")'''

print(f"RMSE score is {rmse}")

print("Metrics saved to model_metrics.txt")
```

### Catboost with target encoding

```python
# Compute mean target encoding only from training data
categorical_cols = ['city_id_merchant_mode', 'state_id_merchant_mode', 'city_id_mode', 'state_id_mode', 'feature_1_mode', 'feature_2_mode', 'feature_3_mode']  # List your categorical columns

# Convert categorical columns to object type to allow mapping
X_train[categorical_cols] = X_train[categorical_cols].astype(str)
X_test[categorical_cols] = X_test[categorical_cols].astype(str)

# Compute target encoding from training data only
category_means_dict = {}  # Store the means for later use

for col in categorical_cols:
    category_means = X_train.join(y_train).groupby(col)['target_mean'].mean()
    category_means_dict[col] = category_means  # Store for debugging if needed

    # Map the mean target value to each category
    X_train[col] = X_train[col].map(category_means)
    X_test[col] = X_test[col].map(category_means)

    # Handle unseen categories in test data
    X_test[col].fillna(y_train.mean(), inplace=True)


```

```python
import pandas as pd
import catboost as cb

# Define CatBoost model
catboost_model = cb.CatBoostRegressor(
    iterations=500, 
    learning_rate=0.1, 
    depth=6, 
    verbose=100
)

# Train the model, specifying categorical features
catboost_model.fit(X_train, y_train)

# Make predictions
y_pred = catboost_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
r2 = r2_score(y_test, y_pred)

# Save metrics to a file
'''with open("model_metrics.txt", "w") as file:
    file.write(f"Mean Absolute Error (MAE): {mae}\n")
    file.write(f"Mean Squared Error (MSE): {mse}\n")
    file.write(f"Root Mean Squared Error (RMSE): {rmse}\n")
    file.write(f"R² Score: {r2}\n")'''

print(f"RMSE score is {rmse}")

print("Metrics saved to model_metrics.txt")
```

### CatBoost k-fold 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import mean_squared_error
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # 5-Fold CV
rmse_scores = []  # Store RMSE for each fold
# Categorical columns
cat_features = ['city_id_merchant_mode', 'state_id_merchant_mode', 'city_id_mode', 'state_id_mode', 'feature_1_mode', 'feature_2_mode', 'feature_3_mode']  
for train_idx, val_idx in kf.split(X):
    # Split data into training and validation sets
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # Create CatBoost Pools (for better performance)
    train_pool = Pool(X_train, y_train, cat_features=cat_features)
    val_pool = Pool(X_val, y_val, cat_features=cat_features)

    # Initialize and train the CatBoost model
    model = CatBoostRegressor(
        iterations=1000,  # Number of trees
        learning_rate=0.05,
        depth=6,
        loss_function='RMSE',  # Using RMSE as the metric
        eval_metric='RMSE',
        cat_features=cat_features,
        verbose=100,
        early_stopping_rounds=50  # Stop if no improvement
    )
    
    model.fit(train_pool, eval_set=val_pool, use_best_model=True)

    # Predict on validation data
    y_pred = model.predict(X_val)

    # Compute RMSE and store the score
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    rmse_scores.append(rmse)
    print(f"Fold RMSE: {rmse:.4f}")

# Print the mean RMSE across all folds
print(f"Mean RMSE across folds: {np.mean(rmse_scores):.4f}")
```
