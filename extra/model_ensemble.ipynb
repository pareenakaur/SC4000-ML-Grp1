```python
combined_df=pd.DataFrame
combined_df=aarushi.merge(full_model,on='card_id').merge(jessica,on='card_id').merge(lavanya,on='card_id').merge(pareena,on='card_id').merge(tanvi,on='card_id')
combined_df
```

```python
combined_df=combined_df.rename(columns={'target':'aarushi'})
```

```python
combined_df
```

```python
# -------- RMSE BASED ON PUBLIC LEADERBOARD CALCULATIONS --------------
# tanvi (lightGBM) : 3.79
# aarushi (lightgbm) :3.87
# jessica (catboost) : 3.92
# lavanya (lightgbm) : 3.91
# pareena (lightgbm) : 3.82
# full model (lightgbm) : 3.76

#Calculate weights based on percentage of inverse RMSE (normalised)
weights = {
    'tanvi':   0.1658,
    'aarushi': 0.1625,
    'jessica': 0.1605,
    'lavanya': 0.1609,
    'pareena': 0.1646,
    'full':    0.1658
}

combined_df['target'] = (
    0.1658 * combined_df['tanvi'] +
    0.1625 * combined_df['aarushi'] +
    0.1605 * combined_df['jessica'] +
    0.1609 * combined_df['lavanya'] +
    0.1646 * combined_df['pareena'] +
    0.1658 * combined_df['full_model']
)

combined_df
```

```python
solution=pd.DataFrame
solution=combined_df.drop(columns=['aarushi','tanvi','pareena','full_model','jessica','lavanya'])
solution
```

```python
pd.DataFrame.to_csv(solution,'solution.csv',index=False)
```
